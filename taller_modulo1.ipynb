### 1.1 — Listas y operaciones básicas

Tienes la siguiente lista con la esperanza de vida de 10 países (en años).

````python
# Lista dada — NO modificar
esperanzas_de_vida = [72.3, 81.5, 60.1, 78.9, 55.4, 83.2, 67.8, 74.0, 49.3, 79.6]
````

**Ejercicio 1.1** — Usando Python puro (sin Pandas ni funciones built-in como `sum()` o `max()`):
- Calcula el **promedio** de la lista usando un bucle `for`
- Encuentra el **valor máximo** usando un bucle `for`
- Encuentra el **valor mínimo** usando un bucle `for`

````python
# Tu código aquí
promedio = 0
num_datos = 0
# Promedio
for i in esperanzas_de_vida:
    promedio +=i
    num_datos +=1


promedio = promedio/num_datos
print(num_datos)
print(f"{promedio:.4}")

# Máximo
x_max = 0
for i in esperanzas_de_vida:
    if x_max < i:
        x_max = i
print(f"valor maximo: {x_max}")

# Mínimo

x_min = x_max
for i in esperanzas_de_vida:
    if x_min > i:
        x_min = i
print(f"valor minimo: {x_min}")

````

### 1.3 — Funciones

Las funciones nos permiten reutilizar lógica.

**Ejercicio 1.3** — Crea una función llamada `clasificar_esperanza_vida` que reciba un número (años) y retorne:
- `"Alta"` si es >= 75
- `"Media"` si es entre 60 y 74.9
- `"Baja"` si es < 60

Prueba la función con al menos 5 valores distintos e imprime los resultados.

````python
# Tu código aquí
def clasificar_esperanza_vida(anios):
    if anios >= 75:
        return "Alta"
    elif 60 < anios < 75:
        return "Media"
    if anios < 60:
        return "Baja"
    
print(clasificar_esperanza_vida(32))
print(clasificar_esperanza_vida(45))
print(clasificar_esperanza_vida(98))
print(clasificar_esperanza_vida(70))
print(clasificar_esperanza_vida(55))
````

### 2.2 — Limpieza de datos

🔧 **Ejercicio 2.2** — El dataset tiene algunos valores nulos (¡datos reales siempre tienen este problema!).  
1. Rellena los valores nulos de columnas **numéricas** con la **media** de cada columna
2. Verifica que ya no queden nulos
3. Explica en una celda de texto: ¿Por qué usamos la media y no simplemente eliminamos esas filas?

````python
for columna in df.columns:
    if df[columna].dtype == 'float64':
        media_columna = df[columna].mean()
        df[columna] = df[columna].fillna(media_columna)
        #print(f"Info limpia de {columna}: {df.shape}")

print(df.isnull().sum())
````

**Responde aquí** (doble clic para editar):

*¿Por qué usamos la media para rellenar nulos en vez de eliminar las filas?*

> Tu respuesta: Porque si eliminara la fila perderia los datos de las otras columnas.

### 2.5 — Agrupaciones

🔧 **Ejercicio 2.5** — Usa `groupby()` para:
1. Calcular la **esperanza de vida promedio** por `region`
2. Calcular el **gasto en salud promedio** por `nivel_ingresos`
3. Calcular el **número de países** por región
4. ¿Qué región tiene la mayor esperanza de vida promedio? ¿Y la menor?

````python
# Tu código aquí

# Esperanza de vida promedio
esperanza_vida_promedio = df.groupby('region')['esperanza_vida'].mean()
print(esperanza_vida_promedio)

# Gasto en salud promedio por nivel ingresos
gatos_salud_promedio = df.groupby('nivel_ingresos')['gasto_salud_usd'].mean()
print(gatos_salud_promedio)

# numero de paises por region
numero_paises = df.groupby('region')["pais"].count()
print(numero_paises)

# Esperanza de vida mayor

promedios_region = df.groupby('region')["esperanza_vida"].mean()

mejor_region = promedios_region.idxmax()
valor_maximo = promedios_region.max()

peor_region = promedios_region.idxmin()
valor_minimo = promedios_region.min()

print(f"La región con mayor esperanza de vida promedio es {mejor_region} con {valor_maximo:.2f} años")
print(f"La región con menor esperanza de vida promedio es {peor_region} con {valor_minimo:.2f} años")

````

**Escribe tu pregunta:**

¿Qué región obtiene la mayor esperanza de vida por cada dólar invertido en salud (eficiencia de gasto)?"

````python
# Tu código aquí

# obtencion de la relacion esperanza de vida vs gasto
df['eficiencia_gasto'] = df['esperanza_vida'] / df['gasto_salud_usd']

# Se hace la agrupacion por region bajo el promedio
eficiencia_por_region = df.groupby('region')['eficiencia_gasto'].mean().sort_values(ascending=False)

# 3. Imprimir el resultado
print("Eficiencia del gasto en salud por región (Años de vida por USD):")
print(eficiencia_por_region)

# 4. Se identifica la region ganadora
ganador = eficiencia_por_region.idxmax()
print(f"\nLa región más eficiente es: {ganador}")
````
